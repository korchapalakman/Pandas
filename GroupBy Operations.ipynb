# Group By
allows us to examine data on a per category basis.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/Users/vazgenosipov/Desktop/Python/Python Data analys/UNZIP_FOR_NOTEBOOKS_FINAL/DATA/mpg.csv") 

df
```

```python
# how has miles per galon changed on average
```

```python
df["model_year"].value_counts()
```

```python
df.groupby("model_year").mean() #return all

# this will group by a category model years
# and will return numeric columns with aggregation
```

```python
df.groupby("model_year").mean()["mpg"]

# just one column
```

```python
# may group by several factors

# this is know as multilevel index
# may take a comparison on years and cylinders

df.groupby(["model_year", "cylinders"]).mean()
```

```python
# this a multilevel index in tuples

df.groupby(["model_year", "cylinders"]).mean().index
```

```python
# may also use describe

df.groupby(["model_year", "cylinders"]).describe()
```

```python
year_cyl = df.groupby(["model_year", "cylinders"]).mean()
```

```python
# may grab by names

year_cyl.index.names
```

```python
# may grab by level

year_cyl.index.levels
```

# OUTside level index (.loc)

```python
# if we are intersted in 70 year
# call below
# be aware! this will drop year column

year_cyl.loc[70]
```

```python
# to select several years
# this will bring back the table model_year
# use list 

year_cyl.loc[[70,82]]
```

# INside level index

```python
year_cyl.loc[(70,4)]

# grab a sindle value
# use a tuple
```

```python
year_cyl.head()
```

## CROSS Secrion

```python
year_cyl.xs(key=70, level = "model_year")

# this will show only year 70 
# just like outer function
```

```python
# if I need instances of only 4 selender across all years
# will exclude the key row
# because they are all the same

year_cyl.xs(key=4, level="cylinders")
```

```python
# sort both outer and inner for maximum prodactivity
# example 6 and 8 for every year

# outer filter
df[df["cylinders"].isin([6,8])]
```

```python
df[df["cylinders"].isin([6,8])].groupby(["model_year", "cylinders"]).mean().head()

# we sort the cylinders out
#
```

```python
df[(df["cylinders"]== 6) | (df["cylinders"]==8)].groupby(["model_year", "cylinders"]).mean().head()

# the same as above without listing of options
```

# SWAP level

```python
year_cyl.swaplevel().head()

# swap group-by indexes
```

# Sort

```python
year_cyl.sort_index(level="model_year", ascending=False).head()

# sort by the most outer level
```

## Specify functions for specific columns

```python
# all colums' rows will perform the aggregaction

df.agg(["std","mean"])
```

```python
# may specify on a specific coloumn

df.agg(["std","mean"])["mpg"]
```

```python
# some aggregate function on some specific columns 
# pass in a dictionary with
# keys - columns 
# values - function calls


df.agg({"mpg": ["max","min"], "weight": ["mean", "std"]})

# will produce NaN on columns you have not passed info
```
