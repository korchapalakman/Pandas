## method #1  Lambda

```python
# take two columns: bill and tip
# we will label tip as genereus depending on tip to tatal bill
```

```python
# name the attributes the same way as columns you intend to pass

def quality (total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
# conversion

                     #select the columns to work   #call lambda on dataframe                  # select right axis
df["Quality_of_tip"] = df[["total_bill", "tip"]].apply(lambda df: quality(df["total_bill"], df["tip"]), axis =1)
                                                           # select x,y in parameters of function
```

```python
df.head()
```

## method #2  - .vectorize

```python
# it transforms functions that are not numpy-aware
# for convinience use - simple sintax


# notise that np.vectorize (funct) - trasformns it ()
# and you pass function parapeters soon after
df["Quality"] = np.vectorize(quality)(df["total_bill"],df["tip"])
```

```python
# this method is more efficient 
# vectorize the function we have created 
# pass it on the columns 
```

```python
df.head()
```
