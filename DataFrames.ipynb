# DataFrames

DataFrames are a group of Series objects that share the same index

```python
import numpy as np
import pandas as pd
```

```python
# create a rondom array

np.random.seed(101)
mydata = np.random.randint(0,101,(4,3))
```

```python
mydata
```

```python
# pre-dataframe information

# 1 create 4 index for 4 rows
my_rows = ["CA", "NY", "AZ", "TX"]

# 2 create 3 index for columns
my_colums = ["Jan", "Feb", "Mar"]
```

```python
# Create a DataFrame

df = pd.DataFrame(mydata)
```

```python
df

# without rows and indexes the colunm would look like this
```

```python
# now lets add rows and columns 

df = pd.DataFrame(mydata, index = my_rows, columns= my_colums)
```

```python
df
```

```python
# check your data frame
# info summary

df.info()
```

```python
# round function 

df["price_per_person"] = np.round(df["total_bill"] / df["size"],2)
```
